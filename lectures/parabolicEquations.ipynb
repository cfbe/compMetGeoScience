# Parabolic equations

## Introduction

Just as with elliptic equations, parabolic equations are a subclass of second-order partial differential equations. All second-order partial differential equations can be written in the form

$$A \frac{\partial^2 f}{\partial x^2} + B \frac{\partial^2 f}{\partial x \partial y} + C \frac{\partial^2 f}{\partial y^2} + D \frac{\partial f}{\partial x} +  E \frac{\partial f}{\partial y} + F f + G = 0$$

A <i>parabolic</i> second-order partial differential equation is an equation on the form above where $B^2-AC = 0$ (remember that elliptic equations were characterized by having $B^2-AC < 0$).

Parabolic equations arise in many physics problems. A typical parabolic equation is the heat equation, also known as the diffusivity equation:

$$D \frac{\partial^2 f}{\partial x^2} = \frac{\partial f}{\partial t}$$

Here $D$ is the diffusivity constant. For the heat equation, $f$ represents the temperature, while $D$ represents the thermal diffusivity. Using the letters in the general equation above, we then have $A = D$ (this $D$ is the diffusivity constant, and not the constant $D$ in the general second-order partial differential equation above) and $E = -1$, yielding $B^2 - AC = 0$.

If $f$ has more than one spatial dimension (and then $\nabla$ is a vector with more than one element), we also consider the equation

$$D \nabla^2 f = \frac{\partial f}{\partial t}$$

a <i>multidimensional parabolic</i> partial differential equation.



## Diffusivity equation

In flow in porous media, the transport is described by Darcy's equation:

$$\vec{q} = -\frac{k}{\mu} \nabla p$$

Here $q$ is the volumetric flow rate, $k$ is the permeability (a measure for how well the porous medium allows for transport of fluids), $\mu$ is the viscosity of the fluid, and $p$ is the fluid pressure.

Assume a volume element $V$ with surface area $A$. The mass rate over the surface $A$ is given by

$$\int_A \rho \vec{q} \cdot \vec{n} dA$$

This mass rate over the surface must equal the negative change in mass inside the volume element, thus

$$\int_A \rho \vec{q} \cdot \vec{n} dA = - \int_V \frac{\partial \rho \phi}{\partial t} dV $$

From the Gauss divergence theorem, we have that

$$\int_A \rho \vec{q} \cdot \vec{n} dA = - \int_V \nabla \cdot \rho \vec{q} dV$$

Thus we get

$$\int_V \nabla \cdot \rho \vec{q} dV = - \int_V \frac{\partial \rho \phi}{\partial t} dV $$

As this holds for any volume $V$, we then have the localized equation

$$\nabla \cdot \rho \vec{q} = - \frac{\partial \rho \phi}{\partial t}$$

Filling in the Darcy equation, and assuming constant permeability and viscosity, we get

$$\frac{k}{\mu} \nabla \cdot (\rho \nabla p) = \frac{\partial (\phi \rho)}{\partial t}$$

The liquid compressibility can be written as

$$c_l = - \frac{1}{V_l} \frac{\partial V_l}{\partial p} = -\frac{\rho}{m}\frac{\partial (m/\rho)}{\partial \rho} \frac{\partial \rho}{\partial p} = -\rho \frac{-1}{\rho^2} \frac{\partial \rho}{\partial p} = \frac{1}{\rho} \frac{\partial \rho}{\partial p}$$

We then have 

$$\nabla\rho \cdot \nabla p = \frac{\partial \rho}{\partial p} \nabla p \cdot  \nabla p   = \rho c_l \nabla p \cdot \nabla p $$

For low compressible fluids, this last term can be disregarded, and we get 

$$\frac{k}{\mu} \nabla \cdot (\rho \nabla p) =   \frac{k}{\mu} \left(\nabla\rho \cdot \nabla p + \rho\nabla^2 p \right) \simeq \frac{k}{\mu} \rho\nabla^2 p $$

Employing the chain rule, we have

$$\begin{align}
\frac{\partial (\rho \phi)}{\partial t} = \rho\frac{\partial \phi}{\partial t} + \phi\frac{\partial \rho}{\partial t} &= \rho\frac{\partial \phi}{\partial p}\frac{\partial p}{\partial t} + \phi\frac{\partial \rho}{\partial p}\frac{\partial p}{\partial t} \\ 
&= \rho \phi \left( \frac{1}{\phi}\frac{\partial  \phi}{\partial p} + \frac{1}{\rho} \frac{\partial \rho}{\partial p} \right) \frac{\partial p}{\partial t}  
\end{align}$$ 

Defining the rock compressibility as

$$c_\phi = \frac{1}{\phi} \frac{\partial \phi}{\partial p}$$

and the total compressibility as the sum of fluid and rock compressibility, $c_t = c_l + c_\phi$, we get

$$\frac{\partial (\rho \phi)}{\partial t} = \rho \phi c_t \frac{\partial p}{\partial t}$$ 

Thus, our equation above can be reformulated as

$$\eta \nabla^2 p = \frac{\partial p}{\partial t}$$

where

$$\eta = \frac{k}{\mu \phi c_t}$$

is called the <i>hydraulic diffusivity</i>, a constant which controls the speed of any pressure disturbance. This equation describes transient (time-dependent) flow. We observe that in the steady-state, where there is no time dependency, we obtain the Laplace equation $\nabla^2 p = 0$, which we have previously considered. Also, observe that the diffusivity equation is of the same form as the heat equation above.


### Analytical solution to 1D equation

Assume now a porous medium with a constant diffusivity constant $\eta$, and where the left and right side pressures are fixed as $p_l$ and $p_r$. 

Using the Fourier series, when can obtain the following equation as a solution to our hydraulic diffusivity equation.

$$\begin{equation}
p(x,t) = \frac{2}{\pi} (p_r-p_l) \sum_{i = 1}^{\infty} \frac{1}{i}  \sin{\left( \frac{i \pi x}{l} \right)} e^{  - \frac{i^2 \pi^2}{l^2} \eta t }  + \frac{p_r-p_l}{l}x + p_l 
\end{equation}$$

By inserting the above equation into the hydraulic diffusivity equation, we can check that it is a solution to the equation. Note that this solution contains an infinite sum (a consequence of being obtained from a Fourier series). When we plot it, we will then sum up to a large number, with <i>large</i> meaning large enough to converge.

Note that the exponent for Euler's number $e$ is scaling with $-t$, thus for large time values $t$, this factor will vanish. Thus, for $t \to \infty$ we will have the following expression for the pressure:

$$p(x,t) = (p_r-p_l)\frac{x}{l} + p_l$$

We see that this is a straight line connecting the two boundary conditions. Note that this is the solution to the elliptic equation $\nabla^2 p = 0$, which we considered in the last lecture.

The Python code below will plot this analytical equation.


```python
import numpy as np
import math
import matplotlib.pyplot as plt

#Creating a class for our problem
class Cparabolic:
    def __init__(self):
        # Length
        self.fL=10.0
        # Left boundary condition
        self.fbcl=1.0
        # Right boundary condition
        self.fbcr=0.0
        # Hydraulic diffusivity [meter squared per second]
        self.fEta=1.0 
        # Number of iterations in analytic solution
        self.inn=1000
        # Number of grid cells
        self.iN=101
        # Array for solution
        self.afx=np.linspace(0.0,self.fL,self.iN)
        # Time step size [seconds]
        self.fDeltat=1.0E-3
        # Spatial grid size
        self.fDeltax=self.fL/(self.iN-1)

    def analytical1DSolution(self,fTime):
        afSum=0.0
        for ii in range(1,self.inn+1):
            afSum+=np.sin(ii*math.pi*self.afx/self.fL)*np.exp(-ii**2*math.pi**2*self.fEta*fTime/(self.fL**2))/ii
        afPressure=self.fbcl+(self.fbcr-self.fbcl)*(self.afx/self.fL+(2.0/math.pi)*afSum)
        return afPressure

tModel=Cparabolic()
```

```python
tModel.afx=np.arange(-20.0,20.0,0.01)

fTime=1.0

for tModel.inn in {0,1,2,5,10}:
    plt.plot(tModel.afx,tModel.analytical1DSolution(fTime),label=str(tModel.inn))

tModel.inn=100
plt.plot(tModel.afx,tModel.analytical1DSolution(fTime),label=str(tModel.inn),color='k',linestyle='dashed')

plt.legend(loc='best')
plt.show()
```

```python
tModel=Cparabolic()

for tTime in {1E-1,1.0,5.0,10.0,100.0}:
    plt.plot(tModel.afx,tModel.analytical1DSolution(tTime),label=str(tTime))

plt.legend(loc='best')
plt.show()
```

## Explicit finite difference

We will now investigate numerical solutions to the diffusivity equation. For this, we will again use finite differences to estimate our derivatives.

The centered finite difference for the second-order derivative is given as

$$\frac{\partial^2 f}{\partial x^2} \simeq \frac{f(x+\Delta x) - 2 f(x) +f(x-\Delta x)}{(\Delta x)^2}$$

For our numerical scheme, we then have

$$\left( \frac{\partial^2 p}{\partial x^2} \right)_i^t = \frac{P_{i+1}^t - 2 P_i^t + P_{i-1}^t}{(\Delta x)^2}$$

The forward finite difference for the first-order derivative gives

$$\left( \frac{\partial p}{\partial t} \right)_i^t = \frac{P_{i}^{t+\Delta t} - P_i^t}{\Delta t}$$

For the partial differential equation 
$$\eta \nabla^2 p = \frac{\partial p}{\partial t}$$
we then use the numerical approximation
$$\eta \left( \frac{\partial^2 p}{\partial x^2} \right)_i^t = \left( \frac{\partial p}{\partial t} \right)_i^t$$
which gives the linear equations

$$\eta \frac{P_{i+1}^t - 2 P_i^t + P_{i-1}^t}{(\Delta x)^2} = \frac{P_{i}^{t+\Delta t} - P_i^t}{\Delta t}$$

We observe that the only unknown in this last equation is $P_{i}^{t+\Delta t}$. Thus, we have an explicit expression for the unknown pressure as

$$P_{i}^{t+\Delta t} = \frac{\eta \Delta t}{(\Delta x)^2} \left(P_{i+1}^t - 2 P_i^t + P_{i-1}^t \right) + P_i^t$$

Note that this method is only stable for

$$\frac{2 \eta \Delta t}{ (\Delta x)^2} \leq 1$$

We can check that we are within the stability criterion here:


```python
print(2*tModel.fEta*tModel.fDeltat/(tModel.fDeltax**2))
```

We can implement the numerical method in Python as follows.

```python
def explicitSolution(self,fTime):
    #Solution vectors
    afPe=np.zeros(self.iN)
    #Set boundary conditions
    afPe[0]=self.fbcl
    afPe[-1]=self.fbcr
    # Create a copy for the numerics
    afPenew=np.copy(afPe)

    #Time counter
    fCurTime=0.0

    while (fCurTime<fTime):
        #Increase the time by one time step (note that we can get a slight error in the final time step)
        fCurTime+=self.fDeltat
        # Calculate the pressures at the next time step from
        # the current pressures
        afPenew[1:-1]=afPe[1:-1]+self.fEta*self.fDeltat/self.fDeltax**2*(afPe[2:]-2*afPe[1:-1]+afPe[:-2])
        afPe=np.copy(afPenew)
    return afPe

# Adding explicit method to class
Cparabolic.explicitSolution=explicitSolution

#Colors for plotting
cmap = plt.get_cmap('gnuplot')
fMaxTime=20.0

for fTime in {1,2,5,10,fMaxTime}:
    plt.plot(tModel.afx,tModel.explicitSolution(fTime),color=cmap(1-fTime/fMaxTime),label=str(fTime))
    plt.plot(tModel.afx,tModel.analytical1DSolution(fTime),linestyle='dashed',color=cmap(1-fTime/fMaxTime))#color='k')#,

plt.legend(loc='best')
plt.show()
```

### Implicit finite difference

We will now investigate the implicit finite difference method.

The backward finite difference for the first-order derivative gives

$$\left( \frac{\partial p}{\partial t} \right)_i^t = \frac{P_{i}^{t} - P_i^{t-\Delta t}}{\Delta t}$$

We can use this for any time step, so substituting $t+\Delta t$ for $t$, we get

$$\left( \frac{\partial p}{\partial t} \right)_i^{t+\Delta t} = \frac{P_{i}^{t+\Delta t} - P_i^{t}}{\Delta t}$$

Note that the right-hand side of this equation is exactly the same as the forward difference.

For the partial differential equation 

$$\eta \nabla^2 p = \frac{\partial p}{\partial t}$$
we then use the numerical approximation

$$\eta \left( \frac{\partial^2 p}{\partial x^2} \right)_i^{t+\Delta t} = \left( \frac{\partial p}{\partial t} \right)_i^{t+\Delta t}$$

which gives the linear equations

$$\eta \frac{P_{i+1}^{t+\Delta t} - 2 P_i^{t+\Delta t} + P_{i-1}^{t+\Delta t}}{(\Delta x)^2} = \frac{P_{i}^{t+\Delta t} - P_i^t}{\Delta t}$$

This equation only has one known, namely $P_i^t$, while the rest are unknowns. Still, we have one equation for each grid cell, and we have in total one unknown for each grid cell, which means we have the same number of equations and unknowns. Thus, we are able to solve this set of linear equations. We will do this by setting up the equations in matrix form.

Let 
$$\alpha = \frac{\eta \Delta t}{(\Delta x)^2}$$
then we have the following set of equations:
$$-\alpha P_{i+1}^{t+\Delta t} + (1+ 2\alpha) P_i^{t+\Delta t} - \alpha P_{i-1}^{t+\Delta t} = P_i^t$$
For the boundary conditions, we can show that we get

$$\begin{align}
(4\alpha +1)P_1^{t + \Delta t} -\frac{4}{3}\alpha P_2^{t + \Delta t}  - \frac{8}{3}\alpha p_l &= P_1^t \\
(4\alpha +1)P_n^{t + \Delta t} -\frac{4}{3}\alpha P_{n-1}^{t + \Delta t}  - \frac{8}{3}\alpha p_r  &= P_n^t
\end{align}$$

We can write this in matrix form as

$$\mathbf{A}\vec{P}^{t+\Delta t} = \vec{P}^t + \vec{b}$$

where $\vec{P}$ is the vector consisting of the pressure values for the different cells, while $\vec{b}$ is the vector containing the boundary conditions.

We then have

$$
\mathbf{A}=
\begin{bmatrix}
4\alpha +1 & -4\alpha/3 & 0 & 0 & \cdots & 0 & 0 & 0 \\
-\alpha & 2\alpha + 1 & -\alpha & 0 & \cdots & 0 & 0 & 0 \\
0 & -\alpha & 2\alpha +1 & -\alpha & \cdots & 0 & 0 & 0 \\
\vdots & \vdots & \vdots & \vdots & \ddots & \vdots & \vdots & \vdots \\
0 & 0 & 0 & 0 & \cdots & 2\alpha +1 & -\alpha & 0 \\
0 & 0 & 0 & 0 & \cdots & -\alpha & 2\alpha +1 & -\alpha \\
0 & 0 & 0 & 0 & \cdots & 0 & -4\alpha/3 & 4\alpha+1 \\
\end{bmatrix} 
$$

$$
\vec{b} =
\begin{bmatrix}
  \frac{8}{3} \alpha p_l \\
  0 \\
  0 \\
  \vdots \\
  0 \\
  0 \\
  \frac{8}{3} \alpha p_r
\end{bmatrix}
$$


Inverting matrix $\mathbf{A}$ then gives the following solution for the pressures:

$$\vec{P}^{t+\Delta t} = \mathbf{A}^{-1}\left(\vec{P}^t + \vec{b}\right)$$

This can be done in Python as follows:


```python
def implicitSolution(self,fTime):
    #Solution vector
    afPe=np.zeros(self.iN-2)

    #Time counter
    fCurTime=0.0

    #Find alpha:
    fAlpha=self.fEta*self.fDeltat/self.fDeltax**2

    #Define A matrix
    aafA=np.zeros((self.iN-2,self.iN-2))
    for ii in range(1,self.iN-3):
        aafA[ii,ii-1]=-fAlpha
        aafA[ii,ii]=2*fAlpha+1
        aafA[ii,ii+1]=-fAlpha
    aafA[0,0]=4*fAlpha+1
    aafA[0,1]=-(4/3)*fAlpha
    aafA[-1,-1]=4*fAlpha+1
    aafA[-1,-2]=-(4/3)*fAlpha

    aafAinv=np.linalg.inv(aafA)

    #Define b vector
    afb=np.zeros(self.iN-2)
    afb[0]=(8/3)*fAlpha*self.fbcl
    afb[-1]=(8/3)*fAlpha*self.fbcr

    while (fCurTime<fTime):
        #Increase the time by one time step (note that we can get a slight error in the final time step)
        fCurTime+=self.fDeltat
        # Calculate the pressures at the next time step from
        # the current pressures
        # Implicit calculation
        afPe=np.dot(aafAinv,afPe+afb)
    return afPe

# Adding explicit method to class
Cparabolic.implicitSolution=implicitSolution

#Colors for plotting
cmap = plt.get_cmap('gnuplot')
fMaxTime=20.0

for fTime in {1,2,5,10,fMaxTime}:
    plt.plot(tModel.afx[1:-1],tModel.implicitSolution(fTime),color=cmap(1-fTime/fMaxTime),label=str(fTime))
    plt.plot(tModel.afx,tModel.explicitSolution(fTime),color=cmap(1-fTime/fMaxTime),linestyle='dotted')
    plt.plot(tModel.afx,tModel.analytical1DSolution(fTime),color=cmap(1-fTime/fMaxTime),linestyle='dashed')

plt.legend(loc='best')
plt.show()
```

We observe that the implicit solution is slightly worse than the explicit solution. This is likely due to more numerical diffusion in the implicit method than in the explicit. The implicit method is typically preferable anyway, as it does not have the same stability criterion as the explicit method; it is stable for all grid sizes. This makes the implicit method more sturdy.

### Error analysis

We have used the finite difference approximation

$$\eta \left( \frac{\partial^2 p}{\partial x^2} \right)_i^t = \left( \frac{\partial p}{\partial t} \right)_i^t$$

or the shifted version for the implicit calculations. From the derivation of finite differences, we know that

$$\frac{p(x+\Delta x) - 2 p(x) +p(x-\Delta x)}{(\Delta x)^2} = \frac{\partial^2 p}{\partial x^2} + \frac{\partial^4 p}{\partial x^4} \frac{1}{12} (\Delta x)^2$$

$$\frac{p(t+\Delta t) - p(t) }{\Delta t} = \frac{\partial p}{\partial t} + \frac{\partial^2 p}{\partial t^2} \frac{1}{2} \Delta t$$

Thus, the error is of order $\mathcal{O}((\Delta x)^2)$ and $\mathcal{O}(\Delta t)$.
